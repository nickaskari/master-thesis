```python
df_filtered = df_filtered[df_filtered['DATA_TYPE_FM'].str.match(r'^SR_([0-9]|10)Y$')]
```

```python
df_filtered = pd.concat([
    df_filtered[df_filtered['DATA_TYPE_FM'] != 'SR_10Y'],
    df_filtered[df_filtered['DATA_TYPE_FM'] == 'SR_10Y']
], ignore_index=True)
```

```python
df_filtered.head()
```

```python
df_pivoted = df_filtered.pivot(index='TIME_PERIOD', columns='DATA_TYPE_FM', values='OBS_VALUE')
```

```python
df_pivoted.head()
```

```python
def maturity_key(col):
    return int(col.split('_')[1][:-1])  # removes 'Y' from 'SR_1Y'

df_pivoted = df_pivoted.reindex(columns=sorted(df_pivoted.columns, key=maturity_key))
```

```python
df_pivoted.head()
```

```python
dates_to_plot = ['2024-12-30', '2004-09-06', '2020-03-23']  
```

```python
df_pivoted.index = pd.to_datetime(df_pivoted.index)
df_subset = df_pivoted.loc[df_pivoted.index.isin(pd.to_datetime(dates_to_plot))]
```

```python
fig = go.Figure()

for date in df_subset.index:
    fig.add_trace(go.Scatter(
        x=df_subset.columns,
        y=df_subset.loc[date],
        mode='lines+markers',
        name=date.strftime('%Y-%m-%d')
    ))

fig.update_layout(
    title="Yield Curves for Selected Dates",
    xaxis_title="Maturity",
    yaxis_title="OBS_VALUE",
    template="plotly_white"
)

fig.show()
```
